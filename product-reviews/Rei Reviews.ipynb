```python
from openai import OpenAI
from config import OPENAI_API_KEY
import os

# Set the API key
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY

# Create the OpenAI client
client = OpenAI()
```

```python
def review_themes(review_json_string):
    return client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "system",
                "content": 'You will be provided a product review in the following json format: {"Product Name": "REI Co-op Skyward 4 Tent", "Review Title": "Not for windy areas", "Rating": 2, "Review": "This tent has many great features. Roomy, can stand up inside, well-designed vestibule area. However, the very small rain fly flaps terribly and very loudly in 30 to 40 mile winds. The fly flaps up and down away from the tent roof. We would have gotten soaked by rain flying in if there had been a rain storm at the same time as the wind. I camp in windy areas frequently, so this isn\'t the tent for me."} you will categorize the themes related to the review outputted in valid json format like so: {"themes":["theme1", "theme2","theme3"]}'
            },
            
            {
                "role": "user",
                "content": review_json_string
            }
            
        ],
        temperature=1,
        max_tokens=256,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

```

```python
review_text = [json.dumps(x) for x in clean_reviews if x['Rating'] < 3 ][1]

review_text
```

```python
open_ai_response = review_themes(review_text)
open_ai_response.choices[0].message.content
```
